```python
class Event:
    __events = {}
    __ID = 0
    def __init__(self,name):
        self.__id = Event.__ID
        Event.__ID+=1
        self.__name = name
        self.__people = []
        self.__active = True
        Event.__events[self.__ID] = self

    def add_person(self,person):
        people = [v.__people if v.__active==True else None for k,v in Event.__events.items()]

        people2 = []
        for each in people:
            if each!=None:
                for each2 in each:
                    people2.append(each2)

        if isinstance(person,People) and person not in people2:
            self.__people.append(person)
        else:
            raise Exception('Person already in an event!')


    @property
    def get_people(self):
        return self.__people
        
class People:
    __peoples = {}
    __ID = 0
    def __init__(self,name):
        self.__id = People.__ID
        People.__ID+=1
        self.__name = name
        People.__peoples[self.__ID] = self

    def _repr_(self):
        print=self.__name
        return print

class Transaction:
    __transactions = {}
    __ID = 0
    def __init__(self,name,amount,payee):

        if name.isalpha()!=True:
            raise TypeError('Incorrect Name!')
        try:
            amount = float(amount)
        except:
            raise TypeError('Incorrect Amount!')

        if isinstance(payee,People)!=True:
            raise TypeError('Incorrect Person!')

        self.__id = Transaction.__ID
        Transaction.__ID += 1
        self.__name = name
        self.__amount = amount
        self.__payee = payee

class KITTY:
    __kittys = {}
    __ID = 0
    def __init__(self,event):
        if isinstance(event,Event)!=True:
            raise TypeError('Incorrect Event!')
        self.__id = KITTY.__ID
        KITTY.__ID+=1
        self.__event = event
        self.__transactions = []
        self.__total_cost = 0
        self.__paid = {k:0 for k in event.get_people}
        self.__balance = {k:0 for k in event.get_people}
        self.__people = self.__event.get_people

    def add_transaction(self,name,amount,payee):
        if payee in self.__people:
            transaction = Transaction(name,float(amount),payee)
            self.__total_cost+=float(amount)
            self.__paid[payee]+=float(amount)
            self.__transactions.append(transaction)
        else:
            raise Exception('This person does not exist in this event!')
    @property
    def get_people_who_paid(self):
        paid = []
        for k,v in self.__paid.items():
            if v!=0:
                paid.append(k)
        return paid

    def calculate_balance(self):
        equal = self.__total_cost/len(self.__people)

        for k,v in self.__paid.items():
            self.__balance[k] = v-equal


    def calculate_who_pays_who(self):
        to_be_paid = {}
        to_pay = {}
        who_pays_who = []
        for k,v in self.__balance.items():
            if v>0:
                to_be_paid[k]=v
            else:
                to_pay[k] = v

        for k,v in to_be_paid.items():
            paid_back = []
            sum = 0
            value = v
            for a,b in to_pay.items():
                if sum>=value:
                    continue
                else:
                    if -b>=value:
                        paid_back.append((a,k,value))
                        sum += value
                        to_pay[a]+=value
                        to_be_paid[k]=0
                        value=0
                        break
                    else:
                        paid_back.append((a,k,-b))
                        sum +=-b
                        to_pay[a]+=-b
                        to_be_paid[k]-=-b
                        value-=-b
            who_pays_who+=paid_back

        return who_pays_who
```

```python
annie = People('Annie')
sally = People('Sallie')
bill  = People('Bill')
event = Event('Concert')
event.add_person(annie)
event.add_person(sally)
event.add_person(bill)
kitty = KITTY(event)
kitty.add_transaction('Tickets',180,annie)
kitty.add_transaction('Dinner',75,sally)
kitty.add_transaction('Drinks',19,bill)
kitty.add_transaction('Taxt',16,bill)
kitty.calculate_balance()
kitty.calculate_who_pays_who()


    

  
```
